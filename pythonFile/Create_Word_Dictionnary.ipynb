## CREATING WORD DICTIONNARY

```python
import pandas as pd
import re
import string
import nltk
from nltk.tokenize import TweetTokenizer
```

```python
df = pd.read_excel('20230509 TextCorpusTraining - Teco.xlsx')
df
```

```python
def get_words():
    strt=''
    for each in df["text"]:
        strt += ' ' + each
        
    return strt

get_words()

```

```python
def delete_extra_spaces(txt):
    txt = re.sub(' +', ' ', txt)
    return txt

def delete_punct(txt):
    txt = "".join([i for i in txt if i not in string.punctuation])
    txt = " ".join(re.findall(r"\w*[a-zA-z]\w*", txt))
    return txt

def set_lower_case(txt):
    txt = "".join([i.lower() for i in txt if i not in string.punctuation])
    return txt

def tokenizing_nltk(txt):
    if not txt:
        print('Test, cellules vides')
        text = ''
    txt = nltk.tokenize.word_tokenize(txt)
    tweet = TweetTokenizer()  
    return txt

def cleaningDfMonth(dfMonth):

    dfMonthTmp = delete_extra_spaces(dfMonth)
    dfMonthTmpp = delete_punct(dfMonthTmp)
    dfMonthTmppp = set_lower_case(dfMonthTmpp)
    return dfMonthTmppp


#Cleaning des quatre sous-datasets
dfAll = cleaningDfMonth(get_words())
dfAll
```

```python
def get_all_words(dfMonth):
    
    tmp=[]
    for tweet in dfMonth.split():
        
        tmp.append(tweet)
    return tmp



def tokenizing_nltkplus(txt):
    if not txt:
        print('TEST TEST.')
        text = ''
    tweet = TweetTokenizer()
    tweet.tokenize(txt)    
    return txt

allstring = get_all_words(dfAll)


def get_refactored_words(stringwordsMonth):
    tab = []
    for ele in stringwordsMonth:
        tokenizing_nltkplus(ele)
        
        tab.append(ele)
    return tab

tokenAll = get_refactored_words(allstring)
tokenAll
```

```python
nltk.download('stopwords')
stopwords = nltk.corpus.stopwords.words('french')
nltk.download('punkt')
nltk.download('omw-1.4')
```

```python
dfst = pd.read_excel('stopwords-fr.xlsx')

def get_all_stopwords():
    tb=[]
    for stp in dfst["stopwords"]:
        tb.append(stp)
    return tb

stopwords_fr = get_all_stopwords()

dfst
```

```python
stopwords_all = nltk.corpus.stopwords.words('english') + stopwords_fr

text_all = " ".join([i for i in tokenAll])
mots_tokens_all = nltk.tokenize.word_tokenize(text_all)

def convertTextStopWords(motsMonth):
    mots_without_stopwords = [i for i in motsMonth if i not in stopwords_all]
    return mots_without_stopwords

txt_all = convertTextStopWords(mots_tokens_all)

stopwords_all

```

```python
import spacy
from spacy.lang.fr.examples import sentences
nlp = spacy.load("fr_core_news_sm")
```

```python
def spacy_lemma(txtMonth):
    resp = []
    my_string = ''
    for i in range(len(txtMonth)):
        my_string += txtMonth[i]
        if i != len(txtMonth) - 1:
            my_string += ', '
            
    transformed_string=my_string.replace(",","")   
    doc = nlp(transformed_string)
    for token in doc:
        resp.append(token.lemma_)
    
    return resp


all_txt_lemma = spacy_lemma(txt_all)
all_txt_lemma

```

```python
import nltk

from collections import Counter



```

```python
from gensim import corpora, models
```

```python
dictionary = corpora.Dictionary([doc.split() for doc in all_txt_lemma])
corpus = [dictionary.doc2bow(doc.split()) for doc in all_txt_lemma]

# Créer un modèle TF-IDF à partir de votre corpus
tfidf = models.TfidfModel(corpus)

# Utiliser le modèle TF-IDF pour calculer les scores de chaque mot dans le corpus
scores = {dictionary.get(idx): value for doc in corpus for idx, value in tfidf[doc]}

# Trier les scores pour chaque mot en ordre décroissant et conserver les 100 mots les plus fréquents
top_words = sorted(scores.items(), key=lambda x: x[1], reverse=False)[:100]

```

```python
import nltk
from nltk.tokenize import word_tokenize
from nltk.probability import FreqDist


tokens = word_tokenize(get_words())

# Suppression des stopwords et des caractères non alphabétiques
tokens = [token.lower() for token in tokens if token.isalpha() and token.lower() not in stopwords_all]

all_txt_lemma = spacy_lemma(tokens)
# Calcul de la fréquence des mots
freq_dist = FreqDist(all_txt_lemma)

# Affichage des 100 mots les plus fréquents
top_words = freq_dist.most_common(300)


top_words_df = pd.DataFrame(top_words, columns = ["mots", "fréquences"])

top_words_df.to_csv('./exportfile/frequenceMots/20230509 Fréquence d\'occurrence de mots tendances - Transition Energetique_DICTTRAIN_TECO.csv', encoding='utf-8-sig', index=False)
top_words_df
```
